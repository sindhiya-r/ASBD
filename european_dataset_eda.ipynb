```python
# Pie Chart
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from an Excel file
file_path = '/content/Team_Atrributes final data.xlsx'
df = pd.read_excel(file_path)

# List of columns to plot pie charts for
columns_to_plot = [
    'buildUpPlaySpeedClass',
    'buildUpPlayDribblingClass',
    'buildUpPlayPositioningClass',
    'buildUpPlayPassingClass',
    'chanceCreationPassingClass',
    'chanceCreationCrossingClass',
    'chanceCreationShootingClass',
    'chanceCreationPositioningClass',
    'defencePressureClass',
    'defenceAggressionClass',
    'defenceTeamWidthClass',
    'defenceDefenderLineClass'
]

# Create a subplot to display all pie charts
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(20, 15))
axes = axes.flatten()  # Flatten to easily access each axis

# Loop through the list of columns and plot a pie chart for each
for i, column in enumerate(columns_to_plot):
    # Get the counts of each unique class in the column
    class_counts = df[column].value_counts()

    # Plot the pie chart for the current column
    axes[i].pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff6666'])

    # Set the title with additional padding
    axes[i].set_title(f'Distribution of {column}', pad=20)  # 'pad' controls the gap between title and chart

    # Ensure the pie chart is drawn as a circle
    axes[i].axis('equal')

# Adjust layout to add more space between subplots and ensure titles don't overlap
plt.subplots_adjust(hspace=0.4, wspace=0.4)  # Adjust horizontal and vertical spacing
plt.show()
```

```python
# Doughnut Chart
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from an Excel file
file_path = '/content/Team_Atrributes final data.xlsx'
df = pd.read_excel(file_path)

# List of columns to plot doughnut charts for
columns_to_plot = [
    'buildUpPlaySpeedClass',
    'buildUpPlayDribblingClass',
    'buildUpPlayPositioningClass',
    'buildUpPlayPassingClass',
    'chanceCreationPassingClass',
    'chanceCreationCrossingClass',
    'chanceCreationShootingClass',
    'chanceCreationPositioningClass',
    'defencePressureClass',
    'defenceAggressionClass',
    'defenceTeamWidthClass',
    'defenceDefenderLineClass'
]

# Create a subplot to display all doughnut charts
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(20, 15))
axes = axes.flatten()

# Loop through the list of columns and plot a doughnut chart for each
for i, column in enumerate(columns_to_plot):
    # Get the counts of each unique class in the column
    class_counts = df[column].value_counts()

    # Plot the doughnut chart for the current column
    wedges, texts, autotexts = axes[i].pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90,
                                           colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff6666'])

    # Create a white circle at the center to transform the pie into a doughnut
    centre_circle = plt.Circle((0, 0), 0.70, fc='white')
    axes[i].add_artist(centre_circle)

    # Set the title with additional padding
    axes[i].set_title(f'Distribution of {column}', pad=20)

    # Ensure the chart is circular
    axes[i].axis('equal')

# Adjust layout
plt.subplots_adjust(hspace=0.4, wspace=0.4)
plt.show()
```

```python
# Area Chart
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/content/Match.xlsx'
df = pd.read_excel(file_path)

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])

# Group by season and calculate total goals
df['season'] = df['date'].dt.year
goals_per_season = df.groupby('season')[['home_team_goal', 'away_team_goal']].sum()
goals_per_season['total_goals'] = goals_per_season['home_team_goal'] + goals_per_season['away_team_goal']

# Plot area chart for goals per season
plt.figure(figsize=(12, 6))
plt.fill_between(goals_per_season.index, goals_per_season['total_goals'], color="skyblue", alpha=0.4)
plt.plot(goals_per_season.index, goals_per_season['total_goals'], color="Slateblue", alpha=0.6)
plt.title("Total Goals Scored per Season")
plt.xlabel("Season")
plt.ylabel("Total Goals")
plt.show()
```

```python
# Radar Chart
import pandas as pd
from math import pi
import matplotlib.pyplot as plt

df = pd.read_excel('/content/Team_Atrributes final data.xlsx')
attributes = ['buildUpPlaySpeed', 'chanceCreationCrossing', 'defencePressure']
values = [df[attr].mean() for attr in attributes]
values += values[:1]
angles = [n / float(len(attributes)) * 2 * pi for n in range(len(attributes))]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, values)
ax.fill(angles, values, alpha=0.3)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(attributes)
plt.title('Radar Chart')
plt.show()
```

```python
# Scatter Plot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Excel file
file_path = '/content/Player_Atrributes final data.xlsx'  # Replace with your Excel file path
data = pd.read_excel(file_path)

# Function to create scatter plots and print correlation
def create_scatter_plot(data, x, y, color, title):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x=x, y=y, color=color)
    plt.title(title)
    plt.xlabel(x.capitalize())
    plt.ylabel(y.capitalize())
    plt.show()

    # Calculate and print correlation
    correlation = data[[x, y]].corr().iloc[0, 1]
    print(f"Correlation between {x} and {y}: {correlation}\n")

# Scatter Plot 1: Shot Power vs Finishing
create_scatter_plot(data, 'shot_power', 'finishing', "blue", 'Shot Power vs Finishing')

# Scatter Plot 2: Acceleration vs Sprint Speed
create_scatter_plot(data, 'acceleration', 'sprint_speed', "green", 'Acceleration vs Sprint Speed')

# Scatter Plot 3: Dribbling vs Ball Control
create_scatter_plot(data, 'dribbling', 'ball_control', "purple", 'Dribbling vs Ball Control')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data from the Excel file
file_path = '/content/drive/MyDrive/soccer_dataset/preprocessed/Player_Atrributes final data.xlsx'  # Change this to your file path
df = pd.read_excel(file_path, engine='openpyxl')

# Step 2: Select the attribute for analysis (e.g., overall_rating)
attribute = 'overall_rating'

# Step 3: Drop rows with missing values in the chosen attribute
df = df.dropna(subset=[attribute])

# Step 4: Calculate frequency counts for the selected attribute
attribute_counts = df[attribute].value_counts().sort_index()

# Step 5: Calculate the cumulative percentage
cumulative_percentage = attribute_counts.cumsum() / attribute_counts.sum() * 100

# Step 6: Plot the Pareto chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for frequencies
ax1.bar(attribute_counts.index, attribute_counts.values, color='skyblue', label='Frequency')
ax1.set_xlabel(attribute)
ax1.set_ylabel('Frequency', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Line chart for cumulative percentage
ax2 = ax1.twinx()
ax2.plot(attribute_counts.index, cumulative_percentage, color='orange', marker='o', linestyle='-', label='Cumulative %')
ax2.set_ylabel('Cumulative Percentage', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Step 7: Add titles and legends
plt.title(f'Pareto Chart for {attribute}')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Step 8: Show the plot
plt.grid()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load the team attributes dataset
team_data = pd.read_excel('/content/drive/MyDrive/soccer_dataset/preprocessed/Team_Atrributes final data.xlsx')

# Convert the 'date' column to datetime format for proper plotting
team_data['date'] = pd.to_datetime(team_data['date'])

# Sort the data by date
team_data.sort_values('date', inplace=True)

# Select only numeric columns for aggregation
numeric_cols = team_data.select_dtypes(include=['number']).columns
grouped_data = team_data.groupby('date')[numeric_cols].mean()

# Check if any numeric columns are empty after filtering
if grouped_data.empty:
    print("No numeric data available for plotting. Please check the dataset.")
else:
    # Prepare data for the funnel chart
    funnel_data = {
        "stage": ['Chance Creation Shooting', 'Chance Creation Crossing', 'Chance Creation Passing', 'Build-up Play Passing', 'Build-up Play Speed'],
        "value": [
            grouped_data['chanceCreationShooting'].mean(),
            grouped_data['chanceCreationCrossing'].mean(),
            grouped_data['chanceCreationPassing'].mean(),
            grouped_data['buildUpPlayPassing'].mean(),
            grouped_data['buildUpPlaySpeed'].mean()
        ]
    }

    # Sort the stages in descending order to create the funnel effect
    funnel_data["value"], funnel_data["stage"] = zip(*sorted(zip(funnel_data["value"], funnel_data["stage"]), reverse=True))

    # Create the funnel chart
    fig = go.Figure(go.Funnel(
        y=funnel_data["stage"],
        x=funnel_data["value"],
        textinfo="value+percent initial"
    ))

    # Customize the layout
    fig.update_layout(
        title='Team Attributes Funnel Chart',
        xaxis_title='Average Value',
        yaxis_title='Attributes',
        hovermode='x unified'
    )

    # Show the plot
    fig.show()
```
