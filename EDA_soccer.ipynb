# **1. COLUMN CHART - Prefered foot, Attacking Work Rate, Defensive Work Rate on PLAYER ATTRIBUTES**


```python
# Column Chart
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/Player_Atrributes final data.xlsx'
player_df = pd.read_excel(file_path)

plt.figure(figsize=(14, 8))

plt.subplot(1, 3, 1)
player_df['preferred_foot'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Preferred Foot Distribution')
plt.xlabel('Preferred Foot')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
player_df['attacking_work_rate'].value_counts().plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Attacking Work Rate Distribution')
plt.xlabel('Attacking Work Rate')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
player_df['defensive_work_rate'].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Defensive Work Rate Distribution')
plt.xlabel('Defensive Work Rate')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# **2. SCATTER PLOT - Shot power vs Finishing, Acceleration vs Sprint Speed , Dribbling vs Ball control**

```python
# Scatter Plot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/content/Player_Atrributes final data.xlsx'
data = pd.read_excel(file_path)

def create_scatter_plot(data, x, y, color, title):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x=x, y=y, color=color)
    plt.title(title)
    plt.xlabel(x.capitalize())
    plt.ylabel(y.capitalize())
    plt.show()

    correlation = data[[x, y]].corr().iloc[0, 1]
    print(f"Correlation between {x} and {y}: {correlation}\n")

create_scatter_plot(data, 'shot_power', 'finishing', "blue", 'Shot Power vs Finishing')

create_scatter_plot(data, 'acceleration', 'sprint_speed', "green", 'Acceleration vs Sprint Speed')

create_scatter_plot(data, 'dribbling', 'ball_control', "purple", 'Dribbling vs Ball Control')
```

# **5. PIE CHARTS**


```python
# Pie Chart
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/Team_Atrributes final data.xlsx'
df = pd.read_excel(file_path)

columns_to_plot = [
    'buildUpPlayDribblingClass',
    'buildUpPlayPositioningClass',
    'buildUpPlayPassingClass',
]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 15))
axes = axes.flatten()


for i, column in enumerate(columns_to_plot):

    class_counts = df[column].value_counts()

    axes[i].pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff6666'])

    axes[i].set_title(f'Distribution of {column}')

    axes[i].axis('equal')

plt.subplots_adjust(hspace=0.2, wspace=0.2)
plt.show()
```

# **6. DOUGHNUT CHART**

```python
# Doughnut Chart
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/Team_Atrributes final data.xlsx'
df = pd.read_excel(file_path)

columns_to_plot = [
    'chanceCreationPassingClass',
    'chanceCreationCrossingClass',
    'chanceCreationShootingClass',
    'chanceCreationPositioningClass']

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 15))
axes = axes.flatten()

for i, column in enumerate(columns_to_plot):
    class_counts = df[column].value_counts()
    wedges, texts, autotexts = axes[i].pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90,
                                           colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff6666'])
    centre_circle = plt.Circle((0, 0), 0.70, fc='white')
    axes[i].add_artist(centre_circle)
    axes[i].set_title(f'Distribution of {column}')
    axes[i].axis('equal')

plt.subplots_adjust(hspace=0.5, wspace=0.3)
plt.show()
```

# **10. PARETO PLOT**

```python
# Pareto Chart
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/Player_Atrributes final data.xlsx'
df = pd.read_excel(file_path)
attribute = 'overall_rating'
df = df.dropna(subset=[attribute])
attribute_counts = df[attribute].value_counts().sort_index()
cumulative_percentage = attribute_counts.cumsum() / attribute_counts.sum() * 100
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(attribute_counts.index, attribute_counts.values, color='skyblue', label='Frequency')
ax1.set_xlabel(attribute)
ax1.set_ylabel('Frequency', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

ax2 = ax1.twinx()
ax2.plot(attribute_counts.index, cumulative_percentage, color='orange', marker='o', linestyle='-', label='Cumulative %')
ax2.set_ylabel('Cumulative Percentage', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title(f'Pareto Chart for {attribute}')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.grid()
plt.show()
```

# **11. RADAR CHART**

```python
# Radar Chart
import pandas as pd
from math import pi
import matplotlib.pyplot as plt

df = pd.read_excel('/content/Team_Atrributes final data.xlsx')
attributes = ['buildUpPlaySpeed', 'chanceCreationCrossing', 'defencePressure']
values = [df[attr].mean() for attr in attributes]
values += values[:1]
angles = [n / float(len(attributes)) * 2 * pi for n in range(len(attributes))]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, values)
ax.fill(angles, values, alpha=0.3)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(attributes)
plt.title('Radar Chart')
plt.show()
```

# **12. LINE CHART**

```python
# Line Chart
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/Team_Atrributes final data.xlsx'
df = pd.read_excel(file_path)

df['date'] = pd.to_datetime(df['date'], errors='coerce')

if 'buildUpPlaySpeed' in df.columns and 'date' in df.columns:
    df_grouped = df.groupby('date')['buildUpPlaySpeed'].mean()

    plt.figure(figsize=(12, 6))
    plt.plot(df_grouped.index, df_grouped.values, marker='o', color='#66b3ff', linestyle='-', linewidth=2)

    plt.title("Build-Up Play Speed Over Date")
    plt.xlabel("Date")
    plt.ylabel("Average Build-Up Play Speed")

    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("'buildUpPlaySpeed' or 'date' columns not found in the dataset.")
```

# **13. AREA CHART**

```python
# Area Chart
import pandas as pd
import matplotlib.pyplot as plt

team_data = pd.read_excel('/content/Team_Atrributes final data.xlsx')

team_data['date'] = pd.to_datetime(team_data['date'])

build_up_speed_trend = team_data.groupby('date')['defenceAggression'].mean()

plt.figure(figsize=(12, 6))
plt.fill_between(build_up_speed_trend.index, build_up_speed_trend, color='skyblue', alpha=0.6)
plt.plot(build_up_speed_trend.index, build_up_speed_trend, color='blue', linewidth=2)

plt.title("Trend of Build-up Play Speed Over Time", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Build-up Play Speed", fontsize=14)
plt.grid(True)
plt.show()
```
